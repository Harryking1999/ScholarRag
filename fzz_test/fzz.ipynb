```python
from serpapi import google_scholar_search
```

```python
##直接搜索
search = google_scholar_search.GoogleScholarSearch({
    "q": "contrastive decoding", 
    "api_key": "ba30b10feae3f3b298a6664711b92fd3875f3e803a2ab9fc43dee32c55df3c28"
  })
result = search.get_dict()
print(result)
```

```python
print(type(result))
f = open("tmp_fzz_direct_search_res.txt", "w+")
f.write(str(result))
f.close()
```

参数信息：
q: 查询词
api_key: serpApi的key
start: 从第几个文档开始
num: 取多少文档，和start一起使用。默认10，最大是20，也就是说取大于20个文档则需要多次请求
cites: 被引用文档的id，可以和q一起使用，有cites的时候是搜索被cites文档引用的部分


字段信息：
organic_results：正经的搜索结果

参考：https://serpapi.com/google-scholar-api
